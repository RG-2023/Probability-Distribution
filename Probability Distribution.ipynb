# Probability Distribution

## Importing Libraries

```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

## Defining Values 

**Setting "ks" and "xs" values** 

```python
ks = np.arange(50)
xs = np.linspace(-5, 10, 2000)
```

## Probability Mass Function (PMF) for Discrete Distribution 

### Discrete Distribution 
**Binomial distribution:** If the random variable X follows the binomial distribution with parameters $n$ and $p$, we write $X\sim B(n, p)$. In the binomial distribution, the probability of getting *k* successful outcomes in *n* independent Bernoulli experiments/trials with each trial having the success probability *p*, is given by the probability mass function:

$$f(k; n,p)=\Pr(X=k)={\binom {n}{k}}p^{k}(1-p)^{n-k}$$
Where,

$k$ is the total number of "successes" (k = 0, 1, 2,...) or data being examined and f(k) is the probability function

$n$ is the total number of trials

$p$ is the probability of success in a single experiment

$q$ is the probability of "failure" in a single experiment = $1-p$


**Poisson distribution:** If X follows a poisson distribution then it is denoted by $X\sim Pois(\lambda)$, which is characterized by a single parameter, lambda ($λ$). A discrete random variable X is said to have a poisson distribution with $λ$, if it has a probability mass function given by:

$$f(k;\lambda )=\Pr(X{=}k)={\frac {\lambda ^{k}e^{-\lambda }}{k!}}$$
where,

$k$ is the number of occurrence (k = 0, 1, 2,...) or data being examined and f(k) is the probability function

$e$ is the Euler's number(e = 2.718)

$\lambda\$ is the rate parameter or an average/mean rate of occurrences (λ > 0) 

```python
pmf_binom = st.binom.pmf(ks, 50, 0.25)
# n = 50, p = 0.25
plt.bar(ks, pmf_binom, label='Binomial Example (d4 die)', alpha=0.8)

pmf_poisson = st.poisson.pmf(ks, 30)
# λ = 30
plt.bar(ks, pmf_poisson, label='Poisson Example (car crash)', alpha=0.8)

plt.legend()
plt.xlabel('k')
plt.ylabel('Prob');

# Questions: 
# 1. What is the chance that we get ten ones when we roll a 50 D4 in total? 
# 2. If there are in average 30 car crashes per night, what is the likelihood of experiencing 50?**

print(st.binom.pmf(10, 50, 0.25))
print(st.poisson.pmf(50, 30))
```

## Probability Density Function (PDF) for Continuous Distribution 
### Continuous Distribution 
**Uniform distribution:** If a continuous random variable X has a uniform distribution with the parameters $a$ and $b$ then the distribution is denoted as $X\sim U(a,b)$. The probability density function (PDF) of the **continuous uniform distribution** is given by:

$$f(x; a,b)={\begin{cases}{\dfrac {1}{b-a}}&{\text{for }}a\leq x\leq b,\\[8pt]0&{\text{for }}x<a\ {\text{ or }}\ x>b.\end{cases}}$$
where,

$x$ is the value of the variable or data being examined and f(x) the probability function

$a$ is the minimum value (= lower bound)

$b$ is the maximum value (= upper bound)


**Normal (or Gaussian) distribution:** When a random variable X is normally distributed with the two parameters, mean $μ (=loc)$ and standard deviation $σ (=scale)$), we write $X\sim {\mathcal {N}}(\mu ,\sigma ^{2})$. ⁠The PDF of the normal distribution is given by:

$$f(x; \mu, \sigma)={\frac {1}{\sqrt {2\pi \sigma ^{2}}}}e^{-{\frac {(x-\mu )^{2}}{2\sigma ^{2}}}}$$
where,

$x$ is the value of the variable or data being examined and f(x) the probability function


**Exponential distribution:** If a random variable X has an exponential distribution, we write $X \sim Exp(\lambda)$. The PDF of an exponential distribution is given by:

$$f(x;\lambda )={\begin{cases}\lambda e^{-\lambda x}&x\geq 0,\\0&x<0.\end{cases}}$$
where,

$x$ is the value of the variable or data being examined and f(x) the probability function

$\lambda$ is the rate parameter (λ > 0)


However, **we can use the $loc$ and $scale$ parameters** in order to shift and/or scale the exponential distribution, where the rate parameter $λ$ is emulated by $scale$. The shape of the exponential distribution does not change with the change in its location or 'loc' parameter. 


**Student's t-distribution:** Student's t distribution (or simply the t distribution) $t_{v}$ is a continuous probability distribution that generalizes the standard normal distribution. Therefore, it is symmetric around zero and bell-shaped. The PDF of Student's t distribution is given by:

$$
f(x; \nu) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \, \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{x^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$
where,

$x$ is the value of the variable or data being examined and f(x) the probability function

$\nu\$ (the Greek letter 'nu') is the parameter representing the number of degrees of freedom (df)

$\Gamma\$ is the gamma function


**Log-normal distribution:**  A positive random variable X is log-normally distributed (i.e. **X ~ Lognormal(μ, $σ^2$)**), if the natural logarithm of X is normally distributed with mean μ and variance $σ^2$, or if **ln(X) ~ N(μ, $σ^2$)**. The probability density function (PDF) of a log-normal distribution is given by:

$$ f(x; \mu, \sigma) = \frac{1}{x \sigma \sqrt{2\pi}} \exp\left(-\frac{(\ln x - \mu)^2}{2\sigma^2}\right) $$  

or,

$$f(x; s) = \frac{1}{s x \sqrt{2\pi}}\exp\left(-\frac{\log^2(x)}{2s^2}\right)$$

Where,

$x$ is the value of the variable or data being examined and f(x) the probability function

$\mu\$ is the mean of the natural logarithm of the variable x

$\sigma\$ is the standard deviation of the natural logarithm of the variable x

$s$ is the shape parameter

However, lognormal distribution in SciPy accepts three parameters named $shape (s)$, $loc$ and $scale$, where the mean $\mu$ corresponds to $scale$ (i.e, $\mu$ = log(scale) or scale = exp($\mu$)) and the standard deviation $\sigma$ corresponds to shape ($\sigma$ = s). The $loc$ parameter can be used and the 'loc' value doesn't change the shape of the distribution.

**Skew-normal distribution:** The skew normal distribution is a continuous probability distribution that generalises the normal distribution to allow for non-zero skewness. The PDF of the skew-normal distribution with parameter $\alpha\$ is given by:

$$f(x; \alpha)=2\phi (x)\Phi (\alpha x)$$
where,

$x$ is the value of the variable or data being examined and f(x) the probability function

$\alpha\$ is the skewness parameter (or shape parameter)

$\phi\$ is the standard normal PDF

$\Phi\$ is the standard normal cumulative distribution function (CDF)

```python
pdf_uniform = st.uniform.pdf(xs, -4, 10)
# a = -4, b = 10 
plt.plot(xs, pdf_uniform, label='Uniform(-4,6)')

pdf_normal = st.norm.pdf(xs, loc=5, scale=2)
# μ (=loc) = 5 and σ (=scale) = 2
plt.plot(xs, pdf_normal, label='Normal(5,2)')

pdf_exponential = st.expon.pdf(xs, loc=-2, scale=2)
# Here, the rate parameter λ = 1/scale = 1/2 (or, scale = 1/λ). Exponential distribution also accepts the 'loc' parameter, and the 'loc' value doesn't change the shape of the distribution.
plt.plot(xs, pdf_exponential, label='Exponential(0.5)')

pdf_studentt = st.t.pdf(xs, df=1, loc=0, scale=1)
# Here, ν (=df) = 1 (for one-sample t-test), and 'loc' and 'scale' parameters have default values of 0 and 1, respectively.
plt.plot(xs, pdf_studentt, label="Student-t(1)")

pdf_lognorm = st.lognorm.pdf(xs, s=1, loc=0, scale=1)
# Here, the shape parameter s is equal to 1, and 'loc' and 'scale' parameters have default values of 0 and 1, respectively.  
plt.plot(xs, pdf_lognorm, label="Lognorm(1)")

pdf_skewnorm = st.skewnorm.pdf(xs, -6)
# Here, the 'skewness' parameter α = -6    
plt.plot(xs, pdf_skewnorm, label='Skewnorm(-6)')

plt.legend()
plt.ylabel("Prob")
plt.xlabel("x");
```
